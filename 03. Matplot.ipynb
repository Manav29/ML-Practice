```python
x=np.linspace(0,5,100)
y3 = x**3
y2 = x**2

plt.plot(x,y2,color="black",linewidth=2,label="Square")
plt.plot(x,y3,color="blue",linewidth=1,label="Cube")


plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Square and Cubic Functions")

plt.legend()
plt.grid()

plt.text(2,80,"Random Text", fontsize=12)                   #Starts at (2,80)

plt.show()
```

Bubble Chart

```python
year = [2013,2014,2015,2016,2017,2018,2019]
avSalary = [10,12,15,19,22,25,29]
population = [100,120,180,250,300,370,485]

plt.scatter(year,avSalary,s=population)                 #Size of Bubble represents the population
```

```python
color_array = ['blue','green','red','red','green','black','blue']
plt.scatter=(year,avSalary)                               #,c=color_array,s=population)
plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]
y2=[3,6,9]
plt.plot(x,y,alpha=0.2,color="black",label="Smaller Alpha")
plt.plot(x,y2,alpha=0.8,color="black",label="Bigger Alpha")
plt.legend()
plt.show()
```

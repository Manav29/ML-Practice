Filling Using Mean of that Flower Type

```python
import numpy as np

df.iloc[2:4,1:3] = np.nan                  #Dont Use df[][]
df.head()
```

```python
df_setosa = df[df['species'] == 'setosa']
df_setosa.head()
```

```python
mu1 = df_setosa["sepal_width"].mean()

df["sepal_width"].fillna(mu1 , inplace=True)
df.head()
```

```python
mu2 = df_setosa["petal_length"].mean()

df["petal_length"].fillna(mu2 , inplace=True)
df.head()
```

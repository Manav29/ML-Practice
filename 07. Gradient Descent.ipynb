```python
import numpy as np
import pandas as pd
X = np.arange(0,5,0.1)
y = X**2
```

```python
d = pd.DataFrame(X)
d["y"] = y
d.head()

d = np.array(d)
#
```

```python
def step_gradient(points,learning_rate,m,c):                                ##Generic GD - Single m array
    m_slope = 0                                                             ##An Array of Zeros
    c_slope = 0
    M = len(points)
    for i in range(M):                   #Summation over i
        x = points[i,0]
        y = points[i,1]
        m_slope += (-2/M)*(y - m*x -c)*x                                    ##for loop to calculate all mi_slopes
        c_slope += (-2/M)*(y - m*x -c)
    m_new = m - learning_rate*m_slope
    c_new = c - learning_rate*c_slope
    
    return m_new,c_new
```

```python
def cost(points,m,c):
    total_cost = 0
    M = len(points)
    for i in range(M):                   #Summation over i
        x = points[i,0]
        y = points[i,1]
        total_cost = (1/M)*((y-m*x-c)**2)
    return total_cost
```

```python
def gd(points,learning_rate,num_iterations):
    m=0
    c=0
    
    for i in range(num_iterations):
        m,c = step_gradient(points,learning_rate,m,c)
        print('Itearation ' ,i ,'    -        Cost : ' , cost(points,m,c))
    return m,c
```

```python
def run():
                                                                   #Load Data
    data = d
    
    #learning_rate = 0.15            #Cost Increasing           #Decrease alpha(learning Rate)
    learning_rate = 0.01           #Cost Decreasing
    
    #num_iterations = 10000   #alpha is small (alpha=0.001)
    num_iterations = 1000        #See when Cost changes only by a small amount
    
    m,c = gd(data,learning_rate,num_iterations)
    
    print(m,c)
    
    
```

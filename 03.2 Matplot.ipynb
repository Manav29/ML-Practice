<b> <h> Points with different colors </h> </b>

```python
year = [2013,2014,2015,2016,2017,2018,2019]
avSalary = [10,12,15,19,22,25,29]
population = [100,120,180,250,300,370,485]

plt.scatter(year,avSalary,s=population,c=c)    
```

```python
year = [2013,2014,2015,2016,2017,2018,2019]
avSalary = [10,12,15,19,22,25,29]
population = [100,120,180,250,300,370,485]

plt.scatter(year,avSalary,s=population,c=c,edgecolor="black",marker='^')    
```

```python
plt.scatter(year,avSalary,s=population,c=c,edgecolor="black",marker='^')    

for i in range(len(year)):
    plt.text(year[i]+0.1,avSalary[i]+0.1,population[i])
    
```

<h> <b> Pie Charts </b> </h>

```python
sizes = [2,3,5,7]
plt.pie(sizes)
```

```python
colors = ["Blue","Pink", "Red", "Purple"]
plt.pie(sizes,colors=colors)
```

```python
labels = ["Apple", "Banana", "Mango", "Orange"]
plt.pie(sizes,colors=colors,labels=labels)
plt.title("Favourate Fruit of a class of Students")
```

```python
plt.pie(sizes,colors=colors,labels=labels,autopct="%0.2f")
plt.title("Favourate Fruit of a class of Students")
```

```python
plt.pie(sizes,colors=colors,labels=labels,autopct="%0.2f%%")
plt.title("Favourate Fruit of a class of Students")
```

```python
explode = [0.2,0,0,0]

plt.pie(sizes,colors=colors,labels=labels,autopct="%0.2f%%", explode=explode)
plt.title("Favourate Fruit of a class of Students")
```

```python
explode = [1,0,0,0]

plt.pie(sizes,colors=colors,labels=labels,autopct="%0.2f%%", explode=explode)
plt.title("Favourate Fruit of a class of Students")
```

```python
explode = [0.2,0.1,0,0]

plt.pie(sizes,colors=colors,labels=labels,autopct="%0.2f%%", explode=explode)
plt.title("Favourate Fruit of a class of Students")
```

This implies our graph is going anticlockwisw

```python
explode = [0.2,0,0,0]

plt.pie(sizes,colors=colors,labels=labels,autopct="%0.2f%%", explode=explode, counterclock=False)
plt.title("Favourate Fruit of a class of Students")
```

```python
explode = [0,0,0,0]           #Default

plt.pie(sizes,colors=colors,labels=labels,autopct="%0.2f%%", explode=explode, counterclock=False)
plt.title("Favourate Fruit of a class of Students")
```

<h1> <b> Histogram </b> </h1>

```python
a = [1,2,3,4,1,3,4,2,1,1]
plt.hist(a)
```

```python
b = [1,2,1,3,5,4,5,5,4,3,3,10,7,6,7,8,8,20,22,11,14]
plt.hist(b)
```

```python
#Data Clubbed
```

```python
plt.hist(b,edgecolor="black")
```

```python
#Only Specific points labelled on X Axis
b=np.array(b)
xticks = np.arange(b.max()+1)
xticks
```

```python
plt.hist(b)
plt.xticks(xticks)
```

```python
plt.hist(b,edgecolor="black")
plt.xticks(xticks)
```

```python
plt.hist(b,bins=25,edgecolor="black")
plt.xticks(xticks)                             #Width Not exactly 1
```

```python
plt.hist(b,bins=21,edgecolor="black")
plt.xticks(xticks)
```

<h> <b> Width of each Bin = (Max - Min)/Number of Bins </b> </h>

```python
plt.hist(b,bins=[1,3,6,10,15,20],edgecolor="black")
plt.xticks(xticks)
```
